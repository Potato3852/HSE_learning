**Задача 7: Вычисление квадратного корня методом Герона**

Напишите функцию `problem_heron(a, x_0, eps)`, которая вычисляет квадратный корень числа методом Герона.
1) a - число, у которого ищется квадратный корень
2) x_0 - начальное приближение
3) eps - точность, с которой находится квадратный корень. (определяется как модуль разницы между приближением и числом a**(-1/2))
Выводится минимальное число итераций, за которое квадратный корень находится с точностью до eps.

**Метод Герона:**  
Итерационная формула:  $x_{n+1} = \frac{1}{2} \left( x_n + \frac{a}{x_n} \right)$   
где $a$ - число, из которого извлекаем корень, $x_n$ - текущее приближение.

При увеличении количества итераций текущее приближение стремится к корню из числа $a$.

Ввод: 3 числа типа float как 3 отдельные строки (см. пример). Первое число - неотрицательно, а третье - положительно.

*Пример работы:*
```
Введите число: 25
Введите начальное приближение: 5
Введите точность: 0.0001
√25 ≈ 5.000000000000007 (достигнуто за 3 итерации)
```
```
Введите число: 2
Введите начальное приближение: 1
Введите точность: 0.000001
√2 ≈ 1.4142135623746899 (достигнуто за 5 итераций)
```
```
Введите число: 10
Введите начальное приближение: 3
Введите точность: 0.0000001
√10 ≈ 3.162277660168379 (достигнуто за 5 итераций)
```
